```python
import socket
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
targetIp = input("Enter the target IP address: ")

def portscan(port):
    try:
        s.connect(targetIp, port)
        return True
    except:
        return False

if __name__ == "__main__":
    for portNumber in range(81):
        if portscan(portNumber):
            print("Port " + str(portNumber) + " is open!")
        else:
            print("Port " + str(portNumber) + " is closed.")
```
